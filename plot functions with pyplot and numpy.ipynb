```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

a = 1
b = 1
c = -6
# numpy.linspace() возвращает заданное количество одинаково удаленных друг от друга чисел из указанного интервала
# например np.linspace(0,7,10) вернет 10 равноудаленных друг от друга чисел из интервала [0,7]
x = np.linspace(-5,5,100)
y = a*x**2 + b*x + c
plt.plot(x,y,'r')
plt.plot([-3,2],[0,0], 'ro', markersize=10)
plt.grid(color='b') # make a grid
plt.title('parabola')
```

```python
for i in x:
    if a*i**2+b*i+c == 0:
        print(i)
```

```python
x = np.linspace(-3,3,100)
y = (x+2)*(x-1)*(x-2)
plt.plot(x,y,'r--') # 'r--' - red dashed
plt.plot([-2, 1, 2], [0, 0, 0], 'go', markersize=10) # add points that make function into 0
plt.grid(color='g')
```

```python
moscow = np.loadtxt('data/moscow.dat')
murmansk = np.loadtxt('data/murmansk.dat')
novosib = np.loadtxt('data/novosib.dat')
plt.plot(np.linspace(1,12,12), moscow)
plt.plot(np.linspace(1,12,12), murmansk)
plt.plot(np.linspace(1,12,12), novosib)
plt.xlim(1,12)
plt.title('Среднемесячная температура за 2016 год')
plt.xlabel('месяцы')
plt.ylabel('среднемесячная температура в С')
plt.grid()
plt.legend(['Москва','Мурманск','Новосибирск'],loc='best')
```

```python
air = np.loadtxt('data/murmansk.dat')
sea = np.loadtxt('data/murmansk_sea.dat')
plt.subplot(211)
plt.title('разница температур воды и воздуха в Мурманске')
plt.plot(air, 'r', label='температура воздуха')
plt.plot(sea, 'b', label='температура воды')
plt.grid()
plt.legend(loc='best')
plt.xlim(0,11)
plt.ylabel('температура в С')
plt.xticks([])
plt.subplot(212)
plt.plot(air-sea,'ko')
plt.xticks(np.linspace(0,11,12),['янв','фев','март','апр','май','июн','июл','авг','сент','окт','ноя','дек'])
plt.xlim(0,11)
plt.grid()
plt.title('температура воздуха минус температура воды')
```
